```python
#3. Find all houses which has more than 3 bedroom and price less than 50k. Update their price with 60k
```

```python
house_rental['Price']=house_rental[(house_rental.Bedroom > 3 )& (house_rental.Price < 50000)]
```

```python
def fun(r):
    if (r.Bedroom > 3) & (r.Price < 50000):
        return 60000
    else:
        return r.Price
    
house_rental['Price'] = house_rental[['Bedroom','Price']].apply(fun, axis =1)
house_rental.head()
```
