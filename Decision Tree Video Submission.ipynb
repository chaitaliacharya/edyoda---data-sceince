## Understanding Decision Tree

```python
What is decision tree?
A decision tree is a flowchart-like tree structure where an internal node represents feature (or attribute),
the branch represents a decision rule, and each leaf node represents the outcome.
```

### How does this algorithm work?


```python
Step 1: Create the decisions from the data
```

```python
Step 2: Select the decision using statistical approach or Attribute Selection Mesaure
```

```python
Step 3: Make that attribute a decision node and break the dataset into smaller subsets.
```

```python
Step 4: Start tree building by repeating this process recursively for each child until one of the condition will match:
        o All the tuples belong to the same attribute value.
        o There are no more remaining attributes.
```
